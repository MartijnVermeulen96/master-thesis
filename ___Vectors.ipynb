```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def draw(x, y): 
    xPlusy = (abs(x[0]+y[0]),abs(x[1]+y[1]))
    array = np.array([[0, 0, x[0], x[1]], 
                      [0, 0, y[0], y[1]], 
                      [0, 0, xPlusy[0], xPlusy[1]]])
    X, Y, U, V = zip(*array)
    fig, ax = plt.subplots()
    plt.ylabel('Y-axis')
    plt.xlabel('X-axis')
    ax.quiver(X[0], Y[0], U[0], V[0], angles='xy', scale_units='xy',color='y',scale=1, label="Duplicate")
    ax.quiver(X[1], Y[1], U[1], V[1], angles='xy', scale_units='xy',color='b',scale=1, label="Pattern")
    ax.quiver(X[2], Y[2], U[2], V[2], angles='xy', scale_units='xy',color='g',scale=1, label="Duplicate&Pattern")
    ax.set_xlim([0, 10])
    ax.set_ylim([0, 10])
    plt.draw()
    plt.show()
    
draw([4,2],[2,4])
```
